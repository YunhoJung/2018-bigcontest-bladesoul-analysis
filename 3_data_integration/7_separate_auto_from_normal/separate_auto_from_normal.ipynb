# 2분할 -> 16분할

```python
os.chdir("../new_2_divide")
id_person = pd.read_csv("person_id_2.csv")
id_auto = pd.read_csv("auto_id_2.csv")
for i in range(1,9):
    #origin data에 대한 분할
    data = pd.read_csv("train_act_"+str(i)+"wk.csv")
    data_person = pd.merge(data, id_person, on="acc_id", how = "inner")
    data_person.to_csv("train_person_"+str(i)+"wk.csv")
    data_auto = pd.merge(data, id_auto, on = "acc_id", how = "inner")
    data_auto.to_csv("train_auto_"+str(i)+"wk.csv", index = False)
    
    #feature 추가된 data에 대한 분할
    data2 = pd.read_csv("train_"+str(i)+"wk.csv")
    data2_person = pd.merge(data2, id_person, on="acc_id", how = "inner")
    data2_person.to_csv("train_person_feature_"+str(i)+"wk.csv")
    data2_auto = pd.merge(data2, id_auto, on="acc_id", how = "inner")
    data2_auto.to_csv("train_auto_feature_"+str(i)+"wk.csv", index = False)
```
