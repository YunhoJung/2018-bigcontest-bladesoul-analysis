```python
import pandas as pd
import os
```

```python
os.chdir("../data")
tr_pay = pd.read_csv('train/train_payment_pre.csv')
te_pay = pd.read_csv('test/test_payment_pre.csv')
```

```python
# payment_amount 가 0인 행은 지운다.
# activity 데이터와 합치기 위해서 설명변수 payment_week 의 이름을 wk 로 수정한다. 
tr_pay = tr_pay[tr_pay['payment_amount']!=0]
te_pay = te_pay[te_pay['payment_amount']!=0]
tr_pay = tr_pay.rename(columns={'payment_week':'wk'})
te_pay = te_pay.rename(columns={'payment_week':'wk'})
```

```python
tr_pay.to_csv('train_payment_merge.csv', index=False)
te_pay.to_csv('test_payment_merge.csv', index=False)
```
