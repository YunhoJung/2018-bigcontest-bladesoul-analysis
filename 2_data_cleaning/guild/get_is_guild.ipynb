### ID 불러오기

### guild 데이터 불러오기

```python
guild_test=pd.read_csv("../data/test_guild.csv")
guild_train=pd.read_csv("../data/train_guild.csv")
guild_test.head()
```

## guild_id/ guild_member_acc_id
- 길드원 이름 -> 길드원수 파악가능 -> 소형 중형 대형 길드
- 길드원 이름이 acc_id와 같으면 , 해당 id는 길드가입자

- 따라서 members,Member_numbers라는 변수 추가

```python
# ',' 로 묶여있는 guild_members_acc_id 풀기위해
def extract_id(row):
    return row.split(',')

guild_test['members'] = guild_test['guild_member_acc_id'].apply(extract_id)
guild_train['members'] = guild_train['guild_member_acc_id'].apply(extract_id)
```

```python
del guild_test['guild_member_acc_id']
del guild_train['guild_member_acc_id']
```

```python
guild_test
```

```python
# 길드원 수 추가
def num_ids(row):
    return len(row)
guild_test['Member_numbers'] = guild_test['members'].apply(num_ids)
guild_train['Member_numbers'] = guild_train['members'].apply(num_ids)
```

```python
guild_train.head()
```

## Next step
- test_id의 acc_id 가 guild_test['members'] 에 들어있다면, 그 acc_id는 길드가입자
- train도 위와 같다
- 따라서 이를 체크해보자

```python
# 먼저 Pandas구조의 데이터를 dictionary바꾼다
# key 는 길드명이 되고, 이에 따른 value는 members (길드원이름) 이다
guild_test_dict= dict([(i,a) for i, a in zip(guild_test.guild_id,guild_test.members)])
guild_train_dict= dict([(i,a) for i, a in zip(guild_train.guild_id,guild_train.members)])
```

```python
# 리스트 형태의 value 값 확인 
guild_test_dict['001ce79ebfcf328508c8d61c476e5bacb3d84f029b56238a33da72f59f494ac6']
```

```python
members_test = list(guild_test_dict.values())
members_train = list(guild_train_dict.values())

# flatten
members_test1=sum(members_test, [])
members_train1=sum(members_train, [])
# 길드명과 구분없이 모든 길드의 길드원 리스트 획득 
print(len(members_test1))
print(len(members_train1))
```

## test_id ,train_id 확인

```python
test_id['is_guild'] = 0
train_id['is_guild'] = 0
test_id.head()
```

```python
train_id.head()
```

```python
#길드멤버에 해당 아이디 존재하면 is_guild = 1, 없으면 0
for i in range(len(test_id)):
    if test_id['acc_id'].loc[i] in members_test1:
        test_id['is_guild'].loc[i] = 1
    else:
        pass
test_id
```

```python
#길드멤버에 해당 아이디 존재하면 is_guild = 1, 없으면 0
for i in range(len(train_id)):
    if train_id['acc_id'].loc[i] in members_train1:
        train_id['is_guild'].loc[i] = 1
    else:
        pass
train_id
```
