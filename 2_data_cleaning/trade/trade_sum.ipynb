```python
import pandas as pd
import os
```

```python
# trade 데이터에서 각 acc_id 의 각 주 동안 재화별 거래량을 추출한다.
os.chdir("../data")
tr = pd.read_csv('trade/train_trade_pre.csv')
te = pd.read_csv('trade/test_trade_pre.csv')
```

```python
# train, test 데이터를 한꺼번에 정리하기 위해서 합해준다.
total = pd.concat([tr,te]).sort_values(by=['trade_week', 'trade_day', 'trade_time', 'source_acc_id', 'target_acc_id', 'item_type'])
total = total.drop_duplicates()
total
```

```python
# 총 8주 간의 데이터이므로 8개로 분할한다.
tr1 = total[total['trade_week'] == 1]
tr2 = total[total['trade_week'] == 2]
tr3 = total[total['trade_week'] == 3]
tr4 = total[total['trade_week'] == 4]
tr5 = total[total['trade_week'] == 5]
tr6 = total[total['trade_week'] == 6]
tr7 = total[total['trade_week'] == 7]
tr8 = total[total['trade_week'] == 8]
```

```python
# item_type 별로 각 주간 누적 거래량을 구한다.
# sou = source_acc_id가 건낸 거래량을 의미
# tar = target_acc_id가 받은 거래량을 의미
sou1 = tr1.item_amount.groupby([tr1.source_acc_id, tr1.item_type]).sum().unstack('item_type')
tar1 = tr1.item_amount.groupby([tr1.target_acc_id, tr1.item_type]).sum().unstack('item_type')
sou2 = tr2.item_amount.groupby([tr2.source_acc_id, tr2.item_type]).sum().unstack('item_type')
tar2 = tr2.item_amount.groupby([tr2.target_acc_id, tr2.item_type]).sum().unstack('item_type')
sou3 = tr3.item_amount.groupby([tr3.source_acc_id, tr3.item_type]).sum().unstack('item_type')
tar3 = tr3.item_amount.groupby([tr3.target_acc_id, tr3.item_type]).sum().unstack('item_type')
sou4 = tr4.item_amount.groupby([tr4.source_acc_id, tr4.item_type]).sum().unstack('item_type')
tar4 = tr4.item_amount.groupby([tr4.target_acc_id, tr4.item_type]).sum().unstack('item_type')
sou5 = tr5.item_amount.groupby([tr5.source_acc_id, tr5.item_type]).sum().unstack('item_type')
tar5 = tr5.item_amount.groupby([tr5.target_acc_id, tr5.item_type]).sum().unstack('item_type')
sou6 = tr6.item_amount.groupby([tr6.source_acc_id, tr6.item_type]).sum().unstack('item_type')
tar6 = tr6.item_amount.groupby([tr6.target_acc_id, tr6.item_type]).sum().unstack('item_type')
sou7 = tr7.item_amount.groupby([tr7.source_acc_id, tr7.item_type]).sum().unstack('item_type')
tar7 = tr7.item_amount.groupby([tr7.target_acc_id, tr7.item_type]).sum().unstack('item_type')
sou8 = tr8.item_amount.groupby([tr8.source_acc_id, tr8.item_type]).sum().unstack('item_type')
tar8 = tr8.item_amount.groupby([tr8.target_acc_id, tr8.item_type]).sum().unstack('item_type')
```

```python
sou1['acc_id'] = sou1.index.get_level_values('source_acc_id')
tar1['acc_id'] = tar1.index.get_level_values('target_acc_id')
sou2['acc_id'] = sou2.index.get_level_values('source_acc_id')
tar2['acc_id'] = tar2.index.get_level_values('target_acc_id')
sou3['acc_id'] = sou3.index.get_level_values('source_acc_id')
tar3['acc_id'] = tar3.index.get_level_values('target_acc_id')
sou4['acc_id'] = sou4.index.get_level_values('source_acc_id')
tar4['acc_id'] = tar4.index.get_level_values('target_acc_id')
sou5['acc_id'] = sou5.index.get_level_values('source_acc_id')
tar5['acc_id'] = tar5.index.get_level_values('target_acc_id')
sou6['acc_id'] = sou6.index.get_level_values('source_acc_id')
tar6['acc_id'] = tar6.index.get_level_values('target_acc_id')
sou7['acc_id'] = sou7.index.get_level_values('source_acc_id')
tar7['acc_id'] = tar7.index.get_level_values('target_acc_id')
sou8['acc_id'] = sou8.index.get_level_values('source_acc_id')
tar8['acc_id'] = tar8.index.get_level_values('target_acc_id')
sou1 = sou1.reset_index(drop=True)
tar1 = tar1.reset_index(drop=True)
sou2 = sou2.reset_index(drop=True)
tar2 = tar2.reset_index(drop=True)
sou3 = sou3.reset_index(drop=True)
tar3 = tar3.reset_index(drop=True)
sou4 = sou4.reset_index(drop=True)
tar4 = tar4.reset_index(drop=True)
sou5 = sou5.reset_index(drop=True)
tar5 = tar5.reset_index(drop=True)
sou6 = sou6.reset_index(drop=True)
tar6 = tar6.reset_index(drop=True)
sou7 = sou7.reset_index(drop=True)
tar7 = tar7.reset_index(drop=True)
sou8 = sou8.reset_index(drop=True)
tar8 = tar8.reset_index(drop=True)
sou1 = sou1.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar1 = tar1.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou2 = sou2.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar2 = tar2.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou3 = sou3.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar3 = tar3.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou4 = sou4.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar4 = tar4.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou5 = sou5.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar5 = tar5.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou6 = sou6.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar6 = tar6.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou7 = sou7.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar7 = tar7.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
sou8 = sou8.rename(columns={'accessory':'give_accessory', 'costume':'give_costume', 'gem':'give_gem', 'grocery':'give_grocery', 'money':'give_money', 'weapon':'give_weapon'})
tar8 = tar8.rename(columns={'accessory':'rec_accessory', 'costume':'rec_costume', 'gem':'rec_gem', 'grocery':'rec_grocery', 'money':'rec_money', 'weapon':'rec_weapon'})
```

```python
sou1 = sou1.fillna(0)
sou2 = sou2.fillna(0)
sou3 = sou3.fillna(0)
sou4 = sou4.fillna(0)
sou5 = sou5.fillna(0)
sou6 = sou6.fillna(0)
sou7 = sou7.fillna(0)
sou8 = sou8.fillna(0)
tar1 = tar1.fillna(0)
tar2 = tar2.fillna(0)
tar3 = tar3.fillna(0)
tar4 = tar4.fillna(0)
tar5 = tar5.fillna(0)
tar6 = tar6.fillna(0)
tar7 = tar7.fillna(0)
tar8 = tar8.fillna(0)
```

```python
trade_sum1 = pd.merge(sou1, tar1, on='acc_id', how='outer')
trade_sum2 = pd.merge(sou2, tar2, on='acc_id', how='outer')
trade_sum3 = pd.merge(sou3, tar3, on='acc_id', how='outer')
trade_sum4 = pd.merge(sou4, tar4, on='acc_id', how='outer')
trade_sum5 = pd.merge(sou5, tar5, on='acc_id', how='outer')
trade_sum6 = pd.merge(sou6, tar6, on='acc_id', how='outer')
trade_sum7 = pd.merge(sou7, tar7, on='acc_id', how='outer')
trade_sum8 = pd.merge(sou8, tar8, on='acc_id', how='outer')
```
