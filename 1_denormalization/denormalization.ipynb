```python
# 각각의 설명변수에 대해 기존값 복원을 실시한다. 모든 변수에 대해 해당 셀 반복수행.
# name='설명변수'
# min0 와 min1 은 정규화된 전체 데이터중에서 가장 작은값과 그 다음으로 작은 값이다.
# s(표준편차) = (min1,min0 실제값의 차이) / (min1 - min0)
# 실제값의 정보는 게임 플레이 경험, 파워북으로부터 얻었다.
# 아래 'quest_exp'의 경우 min0의 실제값은 0 이고 min1의 실제값은 70 이므로 s = 70 / (min1 - min0) 
# m(평균) = min0실제값 - s * min0
name = 'quest_exp'
min0 = sorted(tr_act[name].unique())[0]
min1 = sorted(tr_act[name].unique())[1]
print(f'0 : {min0}\n1 : {min1}')
real_min0 = 0
real_min1 = 70
s = (real_min1 - real_min0) / (min1 - min0)
m = real_min0 - s * min0
print(f's: {s}\nm: {m}')

# 정규화된 설명변수에 s를 곱하고 m을 더하면 복원이 된다.
tr_sample=(tr_act[name] * s + m)
sorted(tr_sample.unique(),reverse=False)
```

```python
# test 데이터는 같은 표준편차와 평균을 가지므로 그대로 계산한다.
te_sample = (te_act[name] * s + m)
sorted(te_sample.unique(),reverse=True)
```

```python
# 각 설명변수에 맞게 소수점 자리수를 조정한다.
round_num = 0
tr_act[name] = tr_sample.round(round_num)
te_act[name] = te_sample.round(round_num)
```

```python
# acc_id, wk 기준으로 정렬한다.
tr_act = tr_act.sort_values(by=['acc_id','wk'])
te_act = te_act.sort_values(by=['acc_id','wk'])
```

```python
# 시간의 경우 분 단위로 조정한다.
tr_act['play_time'] /= 60
te_act['play_time'] /= 60
tr_act['game_combat_time'] /= 60
te_act['game_combat_time'] /= 60
```

```python
# 복원에 사용할 수 있는 아이템의 종류를 정하기 위해 min값, max값을 확인한다.
# min 실제 값 = 1, max 실제 값 = 8 (거래 창의 크기 = 8칸)
# train 데이터는 gem, costume 사용 가능하다.
tr_minw=min(tr_trade['item_amount'][tr_trade['item_type']=='weapon'])
tr_maxw=max(tr_trade['item_amount'][tr_trade['item_type']=='weapon'])
tr_ming=min(tr_trade['item_amount'][tr_trade['item_type']=='gem'])
tr_maxg=max(tr_trade['item_amount'][tr_trade['item_type']=='gem'])
tr_mina=min(tr_trade['item_amount'][tr_trade['item_type']=='accessory'])
tr_maxa=max(tr_trade['item_amount'][tr_trade['item_type']=='accessory'])
tr_minc=min(tr_trade['item_amount'][tr_trade['item_type']=='costume'])
tr_maxc=max(tr_trade['item_amount'][tr_trade['item_type']=='costume'])
print(f'Weapon      - Min Y : {tr_minw}   Max Y : {tr_maxw}\nGem        - Min Y : {tr_ming}   Max Y : {tr_maxg}\n'
      f'Accessory   - Min Y : {tr_mina}   Max Y : {tr_maxa}\nCostume    - Min Y : {tr_minc}   Max Y : {tr_maxc}')
```

```python
# test 데이터는 gem 만 사용 가능하다.
# train, test 데이터의 표준편차, 평균은 동일하므로 gem을 복원에 사용하기로 결정한다.
te_minw=min(te_trade['item_amount'][te_trade['item_type']=='weapon'])
te_maxw=max(te_trade['item_amount'][te_trade['item_type']=='weapon'])
te_ming=min(te_trade['item_amount'][te_trade['item_type']=='gem'])
te_maxg=max(te_trade['item_amount'][te_trade['item_type']=='gem'])
te_mina=min(te_trade['item_amount'][te_trade['item_type']=='accessory'])
te_maxa=max(te_trade['item_amount'][te_trade['item_type']=='accessory'])
te_minc=min(te_trade['item_amount'][te_trade['item_type']=='costume'])
te_maxc=max(te_trade['item_amount'][te_trade['item_type']=='costume'])
print(f'Weapon      - Min Y : {te_minw}   Max Y : {te_maxw}\nGem        - Min Y : {te_ming}   Max Y : {te_maxg}\n'
      f'Accessory   - Min Y : {te_mina}   Max Y : {te_maxa}\nCostume    - Min Y : {te_minc}   Max Y : {te_maxc}')
```
