```python
import pandas as pd
```

```python
total = pd.read_csv('list/total_fin_0902.csv')
sou = pd.read_csv('list/auto_source_dou.csv')
tar = pd.read_csv('list/auto_target_dou.csv')
hard = pd.read_csv('list/hard_list.csv')
```

```python
act_sou = pd.merge(total, sou, on='acc_id', how='inner')
act_tar = pd.merge(total, tar, on='acc_id', how='inner')
act_hard = pd.merge(total, hard, on='acc_id', how='inner')
```

```python
len(act_hard)
```

```python
# 조건에 맞춰 auto source 계정을 추려낸다.
soures = act_sou[(act_sou['raid'] == 0) &
                 (act_sou['raidlight'] == 0) &
                 (act_sou['skilled'] == 0) &
                 (act_sou['pay_per_week'] == 0) &
                 (act_sou['is_guild'] == 0) &
                 (act_sou['guild_chat_wk'] == 0) &
                 (act_sou['making_wk'] == 0) &
                 (act_sou['give_money'] != 0) &
                 (act_sou['give_grocery'] != 0) &
                 (act_sou['give_accessory'] == 0) &
                 (act_sou['give_costume'] == 0) &
                 (act_sou['give_gem'] == 0) &
                 (act_sou['give_weapon'] == 0) &
                 (act_sou['rec_accessory'] == 0) &
                 (act_sou['rec_costume'] == 0) &
                 (act_sou['rec_gem'] == 0) &
                 (act_sou['rec_weapon'] == 0)]
```

```python
soures = soures[(soures['give_money'] > 100) &
                (soures['combat_time_wk'] > 30)]
```

```python
soures = soures[(soures['solo'] > 1) |
                (soures['light'] > 1) |
                (soures['normal'] > 1)]
```

```python
soures = pd.DataFrame(soures['acc_id'])
soures['ant_id'] = 1
soures['master_id'] = 0
```

```python
# 조건에 맞춰 auto target 계정을 추려낸다.
tarres = act_tar[(act_tar['raid'] == 0) &
                 (act_tar['raidlight'] == 0) &
                 (act_tar['skilled'] == 0) &
                 (act_tar['pay_per_week'] == 0) &
                 (act_tar['is_guild'] == 0) &
                 (act_tar['guild_chat_wk'] == 0) &
                 (act_tar['making_wk'] == 0) &
                 (act_tar['rec_money'] != 0) &
                 (act_tar['rec_grocery'] != 0)]
```

```python
tarres = pd.DataFrame(tarres['acc_id'])
tarres['ant_id'] = 0
tarres['master_id'] = 1
```

```python
auto_list = pd.concat([soures,tarres]).sort_values(by='acc_id')
auto_list['auto_id']=1
auto_list
```

```python
auto_list.to_csv('auto_list.csv', index=False)
```
