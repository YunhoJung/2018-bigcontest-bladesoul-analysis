```python
# trade 데이터로부터 이상값들을 추적한다.
os.chdir("../data")
tr = pd.read_csv('trade/total_trade_pre.csv')
```

```python
# auto 계정들과, 버스기사, 골드 구매자를 구분하기 위한 과정이다.
# source가 target에게 여러가지 재화를 지급한 목록을 구한다. 
sou = pd.DataFrame(tr[tr.duplicated(['trade_week','trade_day','trade_time','source_acc_id','target_acc_id'])]['source_acc_id'].unique())
tar = pd.DataFrame(tr[tr.duplicated(['trade_week','trade_day','trade_time','source_acc_id','target_acc_id'])]['target_acc_id'].unique())
```

```python
sou=sou.rename(columns={0:'acc_id'})
tar=tar.rename(columns={0:'acc_id'})
sou=sou.sort_values(by=['acc_id'])
tar=tar.sort_values(by=['acc_id'])
```

```python
print(f'Number of Source : {len(sou)}\nNumber of Target : {len(tar)}')
```

Auto 분류는 이후 R script에서 이어지며, 현재 스크립트 내에서는 버스기사나 레이드 활동자를 추려낸다.

```python
tar.to_csv('auto_target_dou.csv', index=False)
sou = pd.DataFrame(list(set(sou['acc_id'])-set(tar['acc_id']))).rename(columns={0:'acc_id'}).sort_values(by='acc_id')
sou.to_csv('auto_source_dou.csv', index=False)
```

```python
# 버스기사 또는 레이드 이익잉여금 분배는 다음과 같이 이뤄진다.
# 1명의 Source가 money를 같은 금액으로 다수의 Target과 나눠가진다.
bus = tr[tr.duplicated(['trade_week', 'source_acc_id', 'item_type', 'item_amount'])]
bus
```

```python
bus_or_raid = bus[bus['item_type'] == 'money']['source_acc_id'].unique()
bus_or_raid = pd.DataFrame(bus_or_raid).rename(columns={0:'acc_id'}).sort_values(by=['acc_id'])
bus_or_raid
```

```python
b_or_r_rec = bus[bus['item_type'] == 'money']['target_acc_id'].unique()
b_or_r_rec = pd.DataFrame(b_or_r_rec).rename(columns={0:'acc_id'}).sort_values(by=['acc_id'])
b_or_r_rec
```

```python
b_or_r = pd.DataFrame(list(set(bus_or_raid['acc_id'])|set(b_or_r_rec['acc_id']))).rename(columns={0:'acc_id'}).sort_values(by='acc_id')
b_or_r
```

```python
b_o_r = b_or_r.rename(columns={'acc_id':'source_acc_id'})
alpha = pd.merge(tr, b_o_r, on='source_acc_id', how='inner').sort_values(by=['trade_week','trade_day','trade_time','source_acc_id','target_acc_id'])
alpha
```

```python
# 1000 금 이상 오간 데이터만 확인한다.
beta = alpha.drop_duplicates({'trade_week','trade_day','trade_time','source_acc_id','target_acc_id'},keep=False)
beta = beta[(beta['item_type'] == 'money') 
    & (beta['item_amount'] >= 1000)
   ]
beta
```

```python
len(beta['source_acc_id'].unique())
```

```python
len(beta['target_acc_id'].unique())
```

```python
# 게임에서 레이드 분배금을 나눠가지거나 큰 금액을 거래하는 이들의 목록은 32,320개가 있다.
hard_list = pd.DataFrame(list(set(beta['source_acc_id']) | set(beta['target_acc_id']))).rename(columns={0:'acc_id'})
hard_list = hard_list.sort_values(by='acc_id')
print(len(hard_list))
hard_list
```
