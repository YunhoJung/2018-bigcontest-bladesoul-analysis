```python
import pandas as pd
```

```python
# 고속성장이용권을 사용하여 캐릭터를 생성했을 때 튜토리얼을 완료하면 홍문 퀘스트 경험치 4,500,000을 얻는다.
# 이를 따로 분류한다.
tr = pd.read_csv('merge/train_merge_0911.csv')
te = pd.read_csv('merge/test_merge_0911.csv')
```

```python
tr_res = tr[tr['quest_hongmun'] == 4500000]
te_res = te[te['quest_hongmun'] == 4500000]
```

```python
tr_jump = pd.DataFrame(tr_res['acc_id'].unique()).rename(columns={0:'acc_id'})
te_jump = pd.DataFrame(te_res['acc_id'].unique()).rename(columns={0:'acc_id'})
```

```python
res = pd.concat([tr_jump, te_jump]).sort_values(by='acc_id')
res['jump_id'] = 1
res
```

```python
res.to_csv('jump_list.csv', index=False)
```
