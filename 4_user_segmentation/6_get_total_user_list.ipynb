```python
import pandas as pd
```

```python
total = pd.read_csv('list/total_fin_0902.csv')
auto = pd.read_csv('list/auto_list.csv')
hard = pd.read_csv('list/hard_list_re.csv')
dum = pd.read_csv('list/8610_list.csv')
```

```python
auto_li=pd.DataFrame(list(set(auto.acc_id)-set(dum.acc_id)-set(hard.acc_id))).rename(columns={0:'acc_id'}).sort_values(by='acc_id')
```

```python
auto_re=pd.merge(auto,auto_li,on='acc_id')
auto_re
```

```python
hard_li=pd.DataFrame(list(set(hard['acc_id'])-set(dum.acc_id)-set(auto_re.acc_id))).rename(columns={0:'acc_id'}).sort_values(by='acc_id')
hard_li['hard_id']=1
```

```python
res=pd.concat([auto_re,hard_li,dum],sort=False).fillna(0).sort_values(by='acc_id')
```

```python
res
```

```python
len(res)
```

```python
normal=pd.DataFrame(list(set(total['acc_id'])-set(res['acc_id']))).rename(columns={0:'acc_id'}).sort_values(by='acc_id')
normal['normal_id']=1
```

```python
total_list=pd.concat([res,normal],sort=False).fillna(0).sort_values(by='acc_id')
total_list['person_id']=total_list['hard_id']+total_list['normal_id']
total_list['person_id'].max()
```

```python
total_list['ant_id']=total_list['ant_id'].astype(int)
total_list['master_id']=total_list['master_id'].astype(int)
total_list['auto_id']=total_list['auto_id'].astype(int)
total_list['hard_id']=total_list['hard_id'].astype(int)
total_list['8610_id']=total_list['8610_id'].astype(int)
total_list['normal_id']=total_list['normal_id'].astype(int)
total_list['person_id']=total_list['person_id'].astype(int)
total_list=pd.DataFrame(total_list,columns=['acc_id','ant_id','master_id','auto_id','normal_id','hard_id','person_id','8610_id'])
```

```python
total_list
```

```python
jump = pd.read_csv('list/jump_list.csv')
```

```python
total_list = pd.merge(total_list, jump, on='acc_id', how='outer').fillna(0)
```

```python
total_list['jump_id']=total_list['jump_id'].astype(int)
total_list
```
